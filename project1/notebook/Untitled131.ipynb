### Train test Split

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('diseases',axis=1)
y = df['diseases']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
print("Before SMOTE:", Counter(y_train))
smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
```

```python
print("After SMOTE:", Counter(y_train))
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Model Building

```python
def evaluate_model(y_true, y_pred, model_name):
    return {
        "Model": model_name,
        "Accuracy": accuracy_score(y_true, y_pred),
        "Precision": precision_score(y_true, y_pred, average='weighted'),
        "Recall": recall_score(y_true, y_pred, average='weighted'),
        "F1 Score": f1_score(y_true, y_pred, average='weighted'),
        "MSE": mean_squared_error(y_true, y_pred),
        "R2 Score": r2_score(y_true, y_pred)
    }
```

## Tensor flow models

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, Flatten, Dense, Dropout, BatchNormalization, LeakyReLU
import numpy as np
import torch
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_squared_error, r2_score
from keras.callbacks import ModelCheckpoint, EarlyStopping
```

### 1.DNN model

```python
dnn = Sequential([
    Dense(256, activation='tanh', input_shape=(x_train.shape[1],)),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(64, activation='relu'),
    Dense(len(np.unique(y_train)), activation='softmax')
])
```

```python
dnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
save_location = '/content/drive/MyDrive/planing computer project'
os.makedirs(save_location, exist_ok=True)
```

```python
dnn_c = ModelCheckpoint(os.path.join(save_location, 'best_dnn_model.h5'), save_best_only=True, monitor='val_loss', mode='min')
dnn_e= EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

cdd = [dnn_c,dnn_e]
```

```python
dnn.fit(x_train, y_train, epochs=30, batch_size = 32, verbose=1, callbacks= cdd)
```

```python
y_prd_dnn = np.argmax(dnn.predict(x_test), axis=1)
```

```python
print("DNN model training is completed. Best model is saved as 'best_dnn_model.h5'.")
```

### 2.CNN model

```python
cnn = Sequential([
    Conv1D(128, kernel_size=5, activation=LeakyReLU(alpha=0.1), input_shape=(x_train.shape[1], 1)),
    BatchNormalization(),
    Conv1D(64, kernel_size=3, activation='relu'),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.3),
    Dense(len(np.unique(y_train)), activation='softmax')
])
```

```python
cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
cnn_c = ModelCheckpoint(os.path.join(save_location, 'best_cnn_model.h5'), save_best_only=True, monitor='val_loss', mode='min')
cnn_e = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
cdc = [cnn_c,cnn_e]
```

```python
cnn.fit(x_train.values.reshape(-1, x_train.shape[1], 1), y_train, epochs=20, batch_size=32, verbose=1, callbacks= cdc)
```

```python
y_prd_cnn = np.argmax(cnn.predict(x_test.values.reshape(-1, x_test.shape[1], 1)), axis=1)
```

```python
print("CNN model training completed. Best model saved as 'best_cnn_model.h5'.")
```

```python
save_location = '/content/drive/MyDrive/planing computer project'
cnn_model_path = os.path.join(save_location, 'best_cnn_model.h5')

# Load the saved CNN model
cnn = load_model(cnn_model_path)

print("CNN model loaded successfully!")
```

```python
tabt = TabNetClassifier(n_d=16, n_a=16, n_steps=5)
tabt.fit(x_train.values, y_train.values, max_epochs=10, batch_size=128)
y_prd_tabt = tabt.predict(x_test.values)
```

```python
torch.save(tabt, os.path.join(save_location, 'tabtransformer_model.pth'))
print("Tabtranfer model saved successfully.")
```

Evaluate model performance

```python
models = [
    ("DNN", y_prd_dnn),
    ("CNN", y_prd_cnn),
    ("TabTransformer", y_prd_tabt)
]
```

```python
results = [evaluate_model(y_test, y_pred, name) for name, y_pred in models]
results_df = pd.DataFrame(results)
print(results_df.sort_values(by='Accuracy', ascending=False))
```

```python

best_model_name = results_df.sort_values(by='Accuracy', ascending=False).iloc[0]['Model']

if best_model_name == "DNN":
    dnn.save(os.path.join(save_location, 'best_model.h5'))
elif best_model_name == "CNN":
    cnn.save(os.path.join(save_location, 'best_model.h5'))
elif best_model_name == "TabTransformer":
    torch.save(tabt, os.path.join(save_location, 'best_model.pth'))
print(f"Best model ({best_model_name}) saved successfully.")
```

```python
def predict_disease(user_symptoms, model, symptoms_dict, df1):
    # Create an input vector of zeros
    input_vector = np.zeros(len(symptoms_dict))

    # Set corresponding symptom indices to 1
    for symptom in user_symptoms:
        if symptom in symptoms_dict:
            input_vector[symptoms_dict[symptom]] = 1
        else:
            print(f"Warning: '{symptom}' not found in symptom dictionary.")

    # Reshape input to match the model's expected format
    input_vector = np.array(input_vector).reshape(1, -1)

    # Predict the disease using the best model
    predicted_disease_index = np.argmax(model.predict(input_vector), axis=1)[0]

    # Retrieve the disease name using index mapping
    predicted_disease = disease_mapping[predicted_disease_index]

    # Retrieve disease details from df1
    disease_info = df1[df1['disease'].str.lower() == predicted_disease.lower()]

    if disease_info.empty:
        return f"No details found for the predicted disease: {predicted_disease}"

    return disease_info[['disease', 'precautions', 'workouts', 'description', 'diets', 'medications']]

```
